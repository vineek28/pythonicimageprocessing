```python
import matplotlib.pyplot as ply
```

```python
img = cv2.imread('C:/Users/vinee/OneDrive/Desktop/opencv_django/hanagaki.jpg')
```

```python
ply.imshow(img)
```

```python
normal = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ply.imshow(normal)
```

```python
def get_image(image, cmap = None):
    ply.axis('off')
    ply.imshow(image, cmap = cmap)
    ply.imshow
```

```python
get_image(normal)
```

```python
#hsv format
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
get_image(hsv)
```

```python
#Grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
get_image(gray, cmap = 'gray')
```

```python
#Blurred
width, height = normal.shape[:2]

if width > 500:
    k = (50, 50)
elif width > 200 and width <= 500:
    k = (25,25)
else:
    k = (10,10)
    
k = (25,25)
blur = cv2.blur(normal, k)
get_image(blur)
```

```python
#Binary - consists of colors that are pure black or pure white
v, binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
get_image(binary, cmap = 'gray')
```

```python
#invert - converts the white area to black and black to white
invert = cv2.bitwise_not(binary)
get_image(invert, cmap = "gray")
```

```python
# Import the image
file_name = "C:/Users/vinee/OneDrive/Desktop/opencv_django/hanagaki.jpg"
  
# Read the image
src = cv2.imread(file_name, 1)
  
# Convert image to image gray
tmp = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
  
# Applying thresholding technique
_, alpha = cv2.threshold(tmp, 0, 255, cv2.THRESH_BINARY)
  
# Using cv2.split() to split channels 
# of coloured image
b, g, r = cv2.split(src)
  
# Making list of Red, Green, Blue
# Channels and alpha
rgba = [b, g, r, alpha]
  
# Using cv2.merge() to merge rgba
# into a coloured/multi-channeled image
dst = cv2.merge(rgba, 4)
  
# Writing and saving to a new image
cv2.imwrite("gfg_white.png", dst)
```
